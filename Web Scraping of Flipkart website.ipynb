```python
import bs4
from bs4 import BeautifulSoup as bs
import requests
import pandas as pd
```

```python
link = "https://www.flipkart.com/search?q=mobiles&as=on&as-show=on&otracker=AS_Query_TrendingAutoSuggest_1_0_na_na_na&otracker1=AS_Query_TrendingAutoSuggest_1_0_na_na_na&as-pos=1&as-type=TRENDING&suggestionId=mobiles&requestId=6b94a5e2-3870-47cc-b095-44c76cd187b9"
```

```python
page = requests.get(link)
```

```python
print(page.content)
```

```python
soup = bs(page.content, 'html.parser')
```

```python
name= soup.find('div', class_="_4rR01T")
print(name.text)
```

```python
rating= soup.find('div', class_="_3LWZlK")
print(rating.text)
```

```python
specifications = soup.find('div', class_="fMghEO")
print(specifications.text)
```

```python
for each in specifications:
    spec= each.find_all('li', class_="rgWa7D")
    print(spec[0].text)
    print(spec[1].text)
    print(spec[2].text)
    print(spec[3].text)
    print(spec[4].text)
    print(spec[5].text)
 
```

```python
price = soup.find('div', class_="_30jeq3 _1_WHN1")
print(price.text)
```

```python
products=[]
prices=[]
ratings=[]
battery=[]
processor=[]
display=[]
camera=[]
memory=[]
```

```python
for data in soup.find_all('div', class_='_3pLy-c row'):
    names=data.find('div', attrs={'class':"_4rR01T"})
    rating=data.find('div', attrs={'class':"_3LWZlK"})
    price=data.find('div', attrs={'class':"_30jeq3 _1_WHN1"})
    specifications = data.find('div', attrs={'class':"fMghEO"})

    for each in specifications:
        col=each.find_all('li', attrs={'class':'rgWa7D'})
        mem=col[0].text
        disp=col[1].text
        cam=col[2].text
        bat=col[3].text
        proc=col[4].text

    products.append(names.text)
    prices.append(price.text)
    ratings.append(rating)
    battery.append(bat)
    processor.append(proc)
    display.append(disp)
    camera.append(cam)
    memory.append(mem)
```

```python
print(len(products))
print(len(prices))
print(len(ratings))
print(len(battery))
print(len(processor))
print(len(display))
print(len(camera))
```
